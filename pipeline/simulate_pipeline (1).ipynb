```python
import pandas as pd
from google.colab import files
import io

# Step 1: Upload CSV manually
uploaded = files.upload()

# Step 2: Read the uploaded CSV
for fn in uploaded.keys():
    df = pd.read_csv(io.BytesIO(uploaded[fn]))

# Step 3: Preview the first 5 rows
print("ðŸ”¹ Dataset Preview:")
display(df.head())

# Step 4: Show data summary
print("ðŸ”¹ Shape of the dataset:", df.shape)
print("ðŸ”¹ Columns:", df.columns)
df.info()
```

```python
# Drop rows with missing values
df = df.dropna()

# Filter: Only include delivered shipments (simulate status filter)
if 'shipment_status' in df.columns:
    df_filtered = df[df['shipment_status'].str.lower() == 'delivered']
else:
    df_filtered = df.copy()  # If no such column exists

# Convert date columns to datetime (simulate data type transformation)
if 'order_date' in df.columns:
    df_filtered['order_date'] = pd.to_datetime(df_filtered['order_date'])

# Sort by date (simulate ordering logic)
if 'order_date' in df_filtered.columns:
    df_filtered = df_filtered.sort_values(by='order_date')

# Display cleaned and filtered data
print("ðŸ”¹ Cleaned & Filtered Data:")
df_filtered.head()
```

```python
df_filtered.to_csv("cleaned_supply_data.csv", index=False)
files.download("cleaned_supply_data.csv")
```

```python
html = df_filtered.to_html()

with open("supply_table.html", "w") as f:
    f.write(html)

files.download("supply_table.html")
```
